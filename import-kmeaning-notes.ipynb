```python
import os
import re

import pandas as pd

from anki.collection import Collection
from jinja2 import Environment, FileSystemLoader
from IPython.display import display, HTML
from pprint import pprint
```

```python
path_to_collection = os.environ.get("PATH_TO_COLLECTION")
col = Collection(path_to_collection)
```

```python
# Kanji meanings module
kmeaning_deck_id = os.environ.get("KMEANING_DECK_ID")
kmeaning_model_id = os.environ.get("KMEANING_MODEL_ID")

kmeaning_model = col.models.get(kmeaning_model_id)
kmeaning_cols = [field['name'] for field in kmeaning_model['flds']]

kmeaning_model.items
kmeaning_data = pd.read_clipboard(
    sep='\t',
    names=kmeaning_cols)

kmeaning_data = kmeaning_data.fillna('')
kmeaning_data.info()

for record in kmeaning_data.to_dict(orient='records'):
    note = col.new_note(kmeaning_model)
    
    for k,v in record.items():
        note[k]= str(v).strip()
    
    if note.duplicate_or_empty():
        continue
    
    print(kmeaning_data)

    # input_ok = None
    input_ok = input("is output ok? [y/n]")
    # input_ok = input_ok == 'y'

    
    if input_ok:
        output = col.add_note( note, int(kmeaning_deck_id))
        print(output)
```
